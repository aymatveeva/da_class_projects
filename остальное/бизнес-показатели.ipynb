```python
for data in (visits, orders, costs):
    data.columns = data.columns.str.lower()
    data.columns = data.columns.str.replace(' ', '_')
    selected_columns = data.select_dtypes(include=['object']).columns
    for column in selected_columns:
        if column != 'device':
            data[column] = pd.to_datetime(data[column])
            print(column,'—', data[column].dtype)
```

```python
# Количество и процент пропусков в столбцах таблицы
def spaces_stat(data):
    spaces_column_names =[]
    spaces_cnt =[]
    spaces_data =[]
    for i in data.columns:
        spaces_column_names.append(i)
    spaces_column_names = pd.DataFrame(spaces_column_names)
    spaces_column_names.columns = ['column_name']
    for j in data.columns:
        spaces_cnt.append(data[j].isna().sum())
    spaces_cnt = pd.DataFrame(spaces_cnt)
    spaces_cnt.columns = ['spaces_cnt']
    spaces_data = pd.concat([spaces_column_names, spaces_cnt], axis=1)
    spaces_data = spaces_data[spaces_data['spaces_cnt']>0].reset_index(drop=True)
    spaces_data['spaces_pct'] =round(spaces_data['spaces_cnt']/len(data)*100,2)
    spaces_data = spaces_data.sort_values('spaces_pct', ascending=False).reset_index(drop=True)
    if len(spaces_data) == 0:
        spaces_data = 'нет пропусков'
    return spaces_data
```

## Метрики

```python
# Построение графиков
def draw(data, kind, data_y=None, data_x=None, ylabel=None, xlabel=None, vmax=None, fmt=None):
    # Гистограмма
    if kind == 'hist':
        # Количество корзин
        def calc_bins (data, data_y):
            if data[data_y].max() < 20:
                return 100
            else: return 50
        # График
        calculated_bins = calc_bins(data, data_y)
        calculated_range = (data[data_y].quantile(0.1),data[data_y].quantile(.99))
        fig = data[data_y].hist(bins=calculated_bins, range=calculated_range)
        # Оси графика
        if (xlabel is None) and (ylabel is None):
            plt.xlabel('Значение', fontsize=14)
            plt.ylabel('Частота', fontsize=14)
            plt.title('Распределение по {}'.format(data_y), fontsize=16)
    # Тепловая карта     
    elif kind == 'heatmap':
        sns.set(style='white')
        plt.figure(figsize=(13, 9))
        fig = sns.heatmap(data, annot=True, linewidths=1, linecolor='gray', fmt=fmt).set_yticklabels(data.index.date) 
    # Остальные графики
    else:
        fig = data.plot(x=data_x, y=data_y, figsize = (20,6), grid = True, kind = kind, legend=False)
        plt.xlabel('%s'%(xlabel), fontsize=14)
        plt.ylabel('%s'%(ylabel), fontsize=14)    
    return fig
```

#### Retention Rate

```python
visits['first_activity_datetime'] = visits.groupby('uid')['start_ts'].transform('min')
```

```python
visits['first_activity_date'] = visits.groupby('uid')['date'].transform('min')
visits['first_activity_month'] = visits['first_activity_date'].astype('datetime64[M]') 
```

```python
visits['cohort_lifetime'] = ((visits['month'] - visits['first_activity_month'])/ np.timedelta64(1, 'M')).round().astype('int')
```

```python
cohorts_visits = visits.groupby(['first_activity_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
```

```python
initial_users_count_visits = \
cohorts_visits[cohorts_visits['cohort_lifetime'] == 0][['first_activity_month', 'uid']].rename(columns={'uid':'cohort_users'})
```

```python
cohorts_visits = cohorts_visits.merge(initial_users_count_visits, on='first_activity_month')
```

```python
cohorts_visits['retention'] = cohorts_visits['uid']/cohorts_visits['cohort_users']
```

```python
retention_pivot_visits = cohorts_visits.pivot_table(
    index='first_activity_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
)
```

```python
draw(retention_pivot_visits.loc[:,1:11], 'heatmap', vmax=0.1, fmt='.1%')
plt.title('Retention Rate', fontsize=16);
```

```python
RR_mean_2 =(retention_pivot_visits[2].mean()*100).round(2)
RR_mean_2
```

#### Средний чек

```python
ttl_mean_receipt = (orders['revenue'].sum()/orders['uid'].count()).round(2)
ttl_mean_receipt
```

```python
orders_daily = orders.groupby('date').agg({'revenue':'sum', 'uid':'count'}).reset_index().\
               rename(columns={'revenue':'orders_sum', 'uid':'orders_cnt'})
orders_daily['mean_order_sum'] = (orders_daily['orders_sum'] / orders_daily['orders_cnt']).round(2)
```

```python
orders_weekly = orders.groupby('week').agg({'revenue':'sum', 'uid':'count'}).reset_index().\
               rename(columns={'revenue':'orders_sum', 'uid':'orders_cnt'})
orders_weekly['mean_order_sum'] = (orders_daily['orders_sum'] / orders_daily['orders_cnt']).round(2)
```

```python
orders_monthly = orders.groupby('month').agg({'revenue':'sum', 'uid':'count'}).reset_index().\
               rename(columns={'revenue':'orders_sum', 'uid':'orders_cnt'})
orders_monthly['mean_order_sum'] = (orders_daily['orders_sum'] / orders_daily['orders_cnt']).round(2)
```

```python
for data, title, data_x in zip([orders_daily, orders_weekly, orders_monthly]
                               ,['orders_daily', 'orders_weekly', 'orders_monthly']
                               , ['date','week', 'month']):
            draw(data, 'line', 'mean_order_sum', data_x, 'Средний чек', 'Период')
            plt.title('%s'%(title), fontsize=16)
```
